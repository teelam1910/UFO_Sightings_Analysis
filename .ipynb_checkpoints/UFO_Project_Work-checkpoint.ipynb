```python
# Import Pandas, Splinter, and BeautifulSoup
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
# Set Up Splinter
browser = Browser('chrome')
```

```python
# Visit The UFO Data Bank Site
url = 'https://nuforc.org/databank/'
browser.visit(url)
```

```python
# Scrape The Site
html = browser.html

# Create A BeautifulSoup Object From The Scraped HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
# Visit The UFO Sightings Index By State/Country Site
state_country_url = 'https://nuforc.org/ndx?id=loc'
browser.visit(state_country_url)
```

```python
# Scrape The Site
html = browser.html

# Create A BeautifulSoup Object From The Scraped HTML
soup_state_country = BeautifulSoup(html, 'html.parser')
```

```python
print(soup_state_country)
```
